```python
from tensorflow.keras.datasets import mnist

((train_x,train_y),(test_x,test_y)) = mnist.load_data()
```

```python
print(train_x.shape)
print(test_x.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

plt.imshow(train_x[0])
plt.show
```

```python
train_x_flatten = np.reshape(train_x, (60000,28*28))/255
test_x_flatten = np.reshape(test_x, (10000,28*28))/255
```

```python
train_y.shape
```

```python
train_y_encoded = np.zeros((60000,10))
train_y_encoded[np.arange(60000),train_y] = 1

test_y_encoded = np.zeros((10000,10))
test_y_encoded[np.arange(10000),test_y] = 1
```

```python
model = tf.keras.Sequential(
    [
    tf.keras.layers.InputLayer(input_shape=784),
    tf.keras.layers.Dense(100,activation="tanh"),
    tf.keras.layers.Dense(10,activation="softmax")]
)
```

```python
model.compile(loss = "categorical_crossentropy", optimizer="SGD",metrics=["accuracy"])
```

```python
model.fit(train_x_flatten,train_y_encoded,epochs=5)
```

```python
model.evaluate(test_x_flatten, test_y_encoded)
```
